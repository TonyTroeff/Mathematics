<div class="alert alert-block alert-info">

</div>

$$
\begin{gather*}
(x-1)^2+(y-\frac{1}{2})^2=25 \quad \implies \quad O(1,\frac{1}{2}) \\\\
\text{Заместваме в уравнението на правата} \\
\implies \frac{1}{2} = \frac{1}{2}\cdot1 \, \blacksquare
\end{gather*}
$$

## Задача 4

<div class="alert alert-block alert-info">
Математическо очакване на случайна величина <br>
$E(X)=x_1p_1+x_2p_2+\dots+x_kp_k$ <br>
<br>
Дисперсия <br>
$D(X)=(x_1-E(X))^2\,p_1+(x_2-E(X))^2\,p_2+\dots+(x_k-E(X))^2\,p_k$
</div>

Първо нека намерим стойността на параметъра $a$. Ще използваме свойството, че $\displaystyle \sum_{i=1}^k p_i = 1$.
$$
\begin{gather*}
a = 1 - 0.4 - 0.3 - 0.1 \\
\implies a = 0.2
\end{gather*}
$$

След това изчисляваме математическото очакване.
$$
\begin{gather*}
E(X) = -1 \cdot 0.4 + \cancel{0 \cdot 0.3} + 1 \cdot 0.2 + 2 \cdot 0.1 \\
\implies E(X) = 0
\end{gather*}
$$

Накрая просто заместваме по формулата, за да изчислим търсената дисперсия.
$$
\begin{gather*}
D(X) = (-1 - 0)^2 \cdot 0.4 + \cancel{(0 - 0)^2 \cdot 0.3} + (1 - 0)^2 \cdot 0.2 + (2 - 0)^2 \cdot 0.1 \\
D(X) = 0.4 + 0.2 + 0.4 \\
\implies D(X) = 1 \, \blacksquare
\end{gather*}
$$

Верният отговор е <span class="alert alert-success">В)</span>.

## Задача 8

$$
\begin{gather*}
\lim_{x \to 3} \frac{2-\sqrt{1+x}}{x-3} = \frac{0}{0} \\
\stackrel{\text{L'Hopital}}{\implies} \lim_{x \to 3} \frac{(2-(1+x)^{\frac{1}{2}})'}{(x-3)'} = \\
= \lim_{x \to 3} \frac{-\frac{1}{2}(1+x)^{-\frac{1}{2}}}{1} = \\
= -\frac{1}{2} \lim_{x \to 3} \frac{1}{(1+x)^{\frac{1}{2}}} = \\
= -\frac{1}{2} \cdot \frac{1}{\sqrt{1+3}} = \\
= -\frac{1}{4} \, \blacksquare
\end{gather*}
$$

## Задача 9

<div class="alert alert-block alert-info">
Сума на безкрайно намаляваща геометрична прогресия, $\lvert q \rvert < 1$ <br>
$\displaystyle S = \frac{a_1}{1-q}$ <br>
<br>
Дисперсия <br>
$D(X)=(x_1-E(X))^2\,p_1+(x_2-E(X))^2\,p_2+\dots+(x_k-E(X))^2\,p_k$
</div>

$$
\begin{gather*}
a_1 = \sin{45^\circ} \\
q = -\sin{45^\circ} \\
\implies S = \frac{\sin{45^\circ}}{1 + \sin{45^\circ}} = \frac{\frac{\sqrt{2}}{2}}{1 + \frac{\sqrt{2}}{2}} = \frac{\sqrt{2}}{2 + \sqrt{2}} = \\
= \frac{\sqrt{2} \cdot (2 - \sqrt{2})}{(2 + \sqrt{2}) \cdot (2 - \sqrt{2})} = \frac{2\sqrt{2} - 2}{2} = \sqrt{2} - 1 \, \blacksquare
\end{gather*}
$$

Верният отговор е <span class="alert alert-success">Б)</span>.

## Задача 10

<div class="alert alert-block alert-info">
Декартово уравнение на права (с ъглов коефициент $k$) <br>
$y=kx+b$ <br>
<br>
Допирателна към графиката на функция $y=f(x)$ в точката с координати $(x_0; f(x_0))$ <br>
$\displaystyle y = f'(x_0)(x-x_0)+f(x_0)$
</div>

$$
\begin{gather*}
f(x)=2x^3-3x-1 \\
\implies f'(x)=6x^2-3
\end{gather*}
$$

След като намерихме производната, можем да заместим по формулата за допирателна. Нека я отбележим с $t$.

$$
\begin{gather*}
f(x_0) = f(1) = 2 \cdot 1 - 3 \cdot 1 - 1 = -2 \\
f'(x_0) = f'(1) = 6 \cdot 1 - 3 = 3 \\
\\
\implies t: \quad y = 3(x-1)-2 = 3x-5 \, \blacksquare
\end{gather*}
$$

<div class="alert alert-block alert-warning">
Ъглов коефициент на права ($k$ в записа на Декартовото уравнение) се нарича $tg$ (тангенсът) на ъгъла, който правата сключва с положителната посока на оста $Ox$.
</div>

Верният отговор е <span class="alert alert-success">Г)</span>.
