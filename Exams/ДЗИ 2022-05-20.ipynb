## Задача 2

<div class="alert alert-block alert-info">
Нормално уравнение на окръжност с център $O(a,b)$ и радиус $r$ <br>
$(x-a)^2+(y-b)^2=r^2$
</div>

За всеки от възможните отговори проверяваме центърът на дадената окръжност лежи на правата с уравнение $y=\frac{1}{2}x$. Верният отговор е <span class="alert alert-success">В)</span>:
$$
\begin{gather*}
(x-1)^2+(y-\frac{1}{2})^2=25 \quad \implies \quad O(1,\frac{1}{2}) \\\\
\text{Заместваме в уравнението на правата} \\
\implies \frac{1}{2} = \frac{1}{2}\cdot1 \, \blacksquare
\end{gather*}
$$

## Задача 4

<div class="alert alert-block alert-info">
Математическо очакване на случайна величина <br>
$E(X)=x_1p_1+x_2p_2+\dots+x_kp_k$ <br>
<br>
Дисперсия <br>
$D(X)=(x_1-E(X))^2\,p_1+(x_2-E(X))^2\,p_2+\dots+(x_k-E(X))^2\,p_k$
</div>

Първо нека намерим стойността на параметъра $a$. Ще използваме свойството, че $\sum_{i=1}^k p_i = 1$.
$$
\begin{gather*}
a = 1 - 0.4 - 0.3 - 0.1 \\
\implies a = 0.2
\end{gather*}
$$

След това изчисляваме математическото очакване.
$$
\begin{gather*}
E(X) = -1 \cdot 0.4 + \cancel{0 \cdot 0.3} + 1 \cdot 0.2 + 2 \cdot 0.1 \\
\implies E(X) = 0
\end{gather*}
$$

Накрая просто заместваме по формулата, за да изчислим търсената дисперсия.
$$
\begin{gather*}
D(X) = (-1 - 0)^2 \cdot 0.4 + \cancel{(0 - 0)^2 \cdot 0.3} + (1 - 0)^2 \cdot 0.2 + (2 - 0)^2 \cdot 0.1 \\
D(X) = 0.4 + 0.2 + 0.4 \\
\implies D(X) = 1 \, \blacksquare
\end{gather*}
$$

Верният отговор е <span class="alert alert-success">В)</span>.

## Задача 8

$$
\begin{gather*}
\lim_{x \to 3} \frac{2-\sqrt{1+x}}{x-3} = \frac{0}{0} \\
\stackrel{\text{L'Hopital}}{\implies} \lim_{x \to 3} \frac{(2-(1+x)^{\frac{1}{2}})'}{(x-3)'} = \\
= \lim_{x \to 3} \frac{-\frac{1}{2}(1+x)^{-\frac{1}{2}}}{1} = \\
= -\frac{1}{2} \lim_{x \to 3} \frac{1}{(1+x)^{\frac{1}{2}}} = \\
= -\frac{1}{2} \cdot \frac{1}{\sqrt{1+3}} = \\
= -\frac{1}{4} \, \blacksquare
\end{gather*}
$$

Верният отговор е <span class="alert alert-success">А)</span>.
