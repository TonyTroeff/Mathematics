## Задача 13

Ако $\displaystyle \lim_{x \to -\infty} f(x) = c_1$ или $\displaystyle \lim_{x \to \infty} f(x) = c_2$, то функцията има хоризонтална асимптота.

$$
\lim_{x \to \infty} \frac{x^2-4}{x^2-1} = \lim_{x \to \infty} \frac{\cancel{x^2} \cdot (1 - \frac{4}{x^2})}{\cancel{x^2} \cdot (1 - \frac{1}{x^2})} = 1
$$

Ако $\displaystyle \lim_{x \to a^-} f(x) = \pm \infty$ или $\displaystyle \lim_{x \to a^+} f(x) = \pm \infty$, то функцията има вертикална асимптота. Проверяваме с точките, в които функцията се прекъсва.

$$
\begin{gather*}
f_D: \{ \lvert x \rvert \ne 1 \} \\
\\
\lim_{x \to -1} f(x) = \lim_{x \to -1} \frac{(-1)^2-4}{(-1)^2-1} = \lim_{x \to -1} \frac{-3}{0} = \infty \\
\lim_{x \to 1} f(x) = \lim_{x \to 1} \frac{1^2-4}{1^2-1} = \lim_{x \to 1} \frac{3}{0} = \infty \\
\end{gather*}
$$

Верният отговор е <span class="alert alert-success">Г)</span>.
